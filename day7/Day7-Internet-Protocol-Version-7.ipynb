```python
def read_input(filename):
    with open(filename) as f:
        ipv7s = []
        for line in f:
            ipv7s.append(line[:-1])
    return ipv7s

def tls_support1(ipv7addr):
    #print "IPv7addr: %s, Length: %s" %(ipv7addr,len(ipv7addr))
    supports_tls = False
    index = 0
    in_abba_area = True
    abba_in_non_abba_area = False
    abba_in_abba_area = False
    while index < len(ipv7addr)-3:
        #print "Entering while loop"
        if not in_abba_area:
            #print "Is in non-abba-area. Index is: %s, char is %s" %(index, ipv7addr[index])
            while ipv7addr[index] != ']' and index < len(ipv7addr)-4:
                if ipv7addr[index] == ipv7addr[index+3] and ipv7addr[index+1] == ipv7addr[index+2] and ipv7addr[index] != ipv7addr[index+2]:
                    abba_in_non_abba_area = True
                    #print "Found %s in non-abba-area!, index is %s, char is %s" %(ipv7addr[index:index+4],index, ipv7addr[index])
                index += 1
            index += 1
            in_abba_area = True
            #print "going to abba-area, index is %s, char is %s" %(index, ipv7addr[index])
        else:
            #print "Is in abba-area. Index is: %s, char is %s" %(index, ipv7addr[index])
            while index < len(ipv7addr)-3 and '[' not in ipv7addr[index:index+3]:
                if ipv7addr[index] == ipv7addr[index+3] and ipv7addr[index+1] == ipv7addr[index+2] and ipv7addr[index] != ipv7addr[index+2]:
                    abba_in_abba_area = True
                    #print "Found %s, index is %s, char is %s" %(ipv7addr[index:index+4],index, ipv7addr[index])
                index += 1
            #print "going to non-abba-area, index is %s, char is %s" %(index, ipv7addr[index])
            in_abba_area = False
            if index < len(ipv7addr)-3 and ipv7addr[index+3] == '[':
                index += 3
    if abba_in_abba_area and not abba_in_non_abba_area:
        supports_tls = True
    return supports_tls

def ssl_support1(ipv7addr):
    #print "IPv7addr: %s, Length: %s" %(ipv7addr,len(ipv7addr))
    supports_ssl = False
    index = 0
    in_supernet = True
    abas = []
    babs = []
    while index < len(ipv7addr)-3:
        #print "Entering while loop"
        if not in_supernet:
            #print "Is in hypernet area. Index is: %s, char is %s" %(index, ipv7addr[index])
            while index < len(ipv7addr)-2 and ']' not in ipv7addr[index:index+2]:
                if ipv7addr[index] == ipv7addr[index+2] and ipv7addr[index] != ipv7addr[index+1]:
                    babs.append(ipv7addr[index:index+3])
                    #print "Found %s in hypernet area!, index is %s, char is %s" %(ipv7addr[index:index+3],index, ipv7addr[index])
                index += 1
            index += 1
            in_supernet = True
            #print "going to supernet area, index is %s, char is %s" %(index, ipv7addr[index])
            if index < len(ipv7addr)-2 and ipv7addr[index+1] == ']':
                index += 2
                #print "Index is %s, char is %s" %(index, ipv7addr[index])
        else:
            #print "Is in supernet area. Index is: %s, char is %s" %(index, ipv7addr[index])
            while index < len(ipv7addr)-2 and '[' not in ipv7addr[index:index+2]:
                if ipv7addr[index] == ipv7addr[index+2] and ipv7addr[index] != ipv7addr[index+1]:
                    abas.append(ipv7addr[index:index+3])
                    #print "Found %s in supernet area, index is %s, char is %s" %(ipv7addr[index:index+3],index, ipv7addr[index])
                index += 1
            #print "going to hypernet area, index is %s, char is %s" %(index, ipv7addr[index])
            in_supernet = False
            if index < len(ipv7addr)-2 and ipv7addr[index+1] == '[':
                index += 2
                #print "Index is %s, char is %s" %(index, ipv7addr[index])
    #print "abas: %s" %abas
    #print "babs: %s" %babs
    for aba in abas:
        for bab in babs:
            checkstr = bab[1]+bab[0]+bab[1]
            if aba == checkstr:
                supports_ssl = True
                print "IPv7addr: %s, Length: %s" %(ipv7addr,len(ipv7addr))
                print "abas: %s" %abas
                print "babs: %s" %babs
                print "aba: %s. bab: %s" %(aba,bab)
    return supports_ssl

ipv7addrs = read_input('ipv7.txt')
num_ipv7_support_tls = 0
num_ipv7_support_ssl = 0
for addr in ipv7addrs:
    if tls_support1(addr):
        num_ipv7_support_tls += 1
    if ssl_support1(addr):
        num_ipv7_support_ssl += 1

print "The number of IPv7 addressses that support TLS is: %s" %num_ipv7_support_tls
print "The number of IPv7 addressses that support SSL is: %s" %num_ipv7_support_ssl
```
