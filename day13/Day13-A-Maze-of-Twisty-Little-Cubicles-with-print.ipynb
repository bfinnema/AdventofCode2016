```python
import copy

class Path:
    def __init__(self, position, pathid):
        self.position = position
        self.position_log = [position]
        self.valid = True
        self.pathid = pathid
        
    def move(self,position):
        self.position = position
        self.position_log.append(position)
        
    def output(self):
        return 1
        
def space(coordinate):
    x = coordinate[0]
    y = coordinate[1]
    v = x*x + 3*x + 2*x*y + y + y*y + odfn
    num_one = str(bin(v)).count('1')
    num_one_even = True
    if num_one%2 != 0:
        num_one_even = False
    return num_one_even

odfn = 1364
init_position = [1,1]
position = [1,1]
paths = []
directions = [[1,0],[0,1],[0,-1],[-1,0]]
dest_reached = False
target_x = 31
target_y = 39
max_x = 50
max_y = 50
pid = 0
global_log = [init_position]

for d in directions:
    x = position[0] + d[0]
    y = position[1] + d[1]
    print x, y, space([x,y])
    if x>=0 and y>=0:
        if space([x,y]):
            p = Path(init_position, pid)
            pid += 1
            p.move([x,y])
            paths.append(p)
            global_log.append([x,y])

print "# Paths: %s" %len(paths)
for path in paths:
    print path.position_log

round_no = 2
while not dest_reached and round_no < 51:
    print ""
    print "Round number: %s" %round_no
    print "Global LOG: %s" %global_log
    round_no += 1
    new_paths = []
    num_new_paths = 0
    for path in paths:
        print "PATH: %s" %path.pathid
        new_paths_exist = False
        if path.valid:
            print path.position_log
            dead_end = True
            freeze_path = copy.deepcopy(path)
            log = freeze_path.position_log
            prev_pos = path.position
            for d in directions:
                x = prev_pos[0] + d[0]
                y = prev_pos[1] + d[1]
                print x, y, space([x,y])
                if x>=0 and x<max_x and y>=0 and y<max_y and [x,y] not in global_log:
                    if space([x,y]):
                        dead_end = False
                        print "This x,y is good: %s, %s, %s" %(x, y, space([x,y]))
                        global_log.append([x,y])
                        if [x,y] == [target_x,target_y]:
                            print "##############################"
                            print "!!!!!!! Reached %s, %s !!!!!!!" %(target_x,target_y)
                            print "##############################"
                            dest_reached = True
                        if new_paths_exist:
                            #print "New path"
                            #print log
                            #new_paths.append(Path(prev_pos, pid))
                            logcopy = copy.deepcopy(log)
                            #new_paths[num_new_paths].position_log = logcopy
                            #new_paths[num_new_paths].move([x,y])
                            newp = Path(prev_pos, pid)
                            newp.position_log = logcopy
                            newp.move([x,y])
                            new_paths.append(newp)
                            pid += 1
                            print "New path: %s %s" %(new_paths[num_new_paths].pathid, new_paths[num_new_paths].position_log)
                            num_new_paths += 1
                        else:
                            path.move([x,y])
                            print "Existing Path: %s %s" %(path.pathid, path.position_log)
                            new_paths_exist = True
            if dead_end:
                path.valid = False
                print "%s: Path met a dead-end" %path.pathid
        else:
            print "%s: This path earlier met a dead-end and has become invalid" %path.pathid
    if len(new_paths) > 0:
        print "New Paths being appended:"
    for npath in new_paths:
        print "Pathid: %s, LOG: %s" %(npath.pathid, npath.position_log)
        paths.append(npath)
    for path in paths:
        print "PATH: %s Position: %s Validity: %s" %(path.pathid, path.position, path.valid)
    #dest_reached = True

print "# Paths: %s" %len(paths)
for path in paths:
    if path.valid:
        print ""
        print "PATH: %s" %path.pathid
        if [target_x,target_y] in path.position_log:
            print "##############################"
            print "!!!!!!! Reached %s, %s !!!!!!!" %(target_x,target_y)
            print "!!!!!!! in %s Steps !!!!!!!!!!" %(len(path.position_log)-1)
            print "##############################"
        print "Last Position: %s" %path.position
        print "Position LOG: %s" %path.position_log
        print "Number of steps: %s" %(len(path.position_log)-1)
        print path.valid
print ""
print global_log, len(global_log)
```
