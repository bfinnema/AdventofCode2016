```python
def read_instr(instr_str):
    instr = ''
    i = 0
    #print i, instr_str
    while instr_str[i] != ')':
        instr += instr_str[i]
        #print "Letter: %s, instr: %s" %(instr_str[i], instr)
        i += 1
    #print instr
    return [int(instr.split('x')[0]),int(instr.split('x')[1]),len(instr)]

def recursive_decomp(c):
    #print c
    #print len(c)
    original_length = len(c)
    posrec = 0
    decomp = 0
    instr_found = False
    while posrec < len(c):
        if c[posrec] == '(':
            instr_found = True
            decomp_slice = 0
            instr = read_instr(c[posrec+1:])
            posrec += instr[2]+2
            #print "Number of characters: %s. Number of repetitions: %s. Position: %s" %(instr[0],instr[1],posrec)
            slice = c[posrec:posrec+instr[0]]
            #print "Calling recursive with %s" %slice
            #print ""
            decomp_slice = recursive_decomp(slice)
            #print "decomp_slice: %s" %decomp_slice
            decomp += instr[1] * decomp_slice
            #print "decomp: %s" %decomp
            posrec += instr[0]
        else:
            posrec += 1
    if not instr_found:
        decomp = original_length
    return decomp

my_file = open('input.txt')
c = my_file.read()[:-1]
#print c
#print len(c)

decompressed = ''
pos = 0

while pos < len(c):
    if c[pos] == '(':
        instr = read_instr(c[pos+1:])
        repstr = ''
        pos += instr[2]+2
        #print "Number of characters: %s. Number of repetitions: %s. Position: %s" %(instr[0],instr[1],pos)
        if pos + instr[0] < len(c):
            #print "addition::  %s" %c[pos:pos+instr[0]]
            for i in range(instr[1]):
                repstr += c[pos:pos+instr[0]]
        else:
            #print "String ending, addition::  %s" %c[pos:len(c)]
            for i in range(instr[1]):
                repstr += c[pos:len(c)]
        #print repstr
        decompressed += repstr
        pos += instr[0]
    else:
        #print "@@@@@@@@@@@@@@@@@@@@@   There is a free character %s" %c[pos]
        decompressed += c[pos]
        pos += 1

print "Part 1: The decompressed length is %s" %len(decompressed)
print"__________________________________________________________________"
print ""

#c = 'X(8x2)(3x3)ABCY'
#c = '(25x3)(3x3)ABC(2x3)XY(5x2)PQRSTX(18x9)(3x2)TWO(5x7)SEVEN'
#c = '(27x12)(20x12)(13x14)(7x10)(1x12)A'
decompressed = 0
pos = 0
#print c
while pos < len(c):
    if c[pos] == '(':
        decomp_slice = 0
        instr = read_instr(c[pos+1:])
        pos += instr[2]+2
        #print "Number of characters: %s. Number of repetitions: %s. Position: %s" %(instr[0],instr[1],pos)
        slice = c[pos:pos+instr[0]]
        #print "Calling recursive with %s" %slice
        #print ""
        decomp_slice = recursive_decomp(slice)
        #print "decomp_slice: %s" %decomp_slice
        decompressed += instr[1] * decomp_slice
        #print "decompressed: %s" %decompressed
        pos += instr[0]
    else:
        #print "@@@@ Free character is %s" %c[pos]
        pos += 1
        decompressed += 1

print "Part 2: The decompressed length is %s" %decompressed
```
